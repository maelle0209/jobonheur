```python
# Reformater les données pour un format "long"
df_long = pd.melt(data, id_vars=["Sexe et ƒge"], var_name="Trimestre", value_name="Taux")

# Séparer le "Sexe et ƒge" en deux colonnes
df_long[['Sexe', 'Âge']] = df_long['Sexe et ƒge'].str.extract(r'(Femmes|Hommes|Ensemble)?(.*)')

# Remplir les lignes "Sexe" manquantes par les valeurs précédentes
df_long['Sexe'] = df_long['Sexe'].fillna(method='ffill')

# Vérifiez le résultat
print(df_long.head())

```

```python
# Vérifier le type de la colonne "Trimestre"
print(df_long.dtypes)

# Si "Trimestre" n'est pas une chaîne, convertissez-la
df_long["Trimestre"] = df_long["Trimestre"].astype(str)
```

```python
import plotly.express as px

# Filtrer les données pour exclure le chômage de longue durée et les tranches indésirables
df_age_filtered = df_long[df_long['Âge'] != "Longue durée"]  # Exclure "Longue durée"
df_age_avg = df_age_filtered.groupby("Âge", as_index=False)['Taux'].mean()

# Exclure la première et la dernière tranche d'âge
df_age_avg_filtered = df_age_avg.iloc[1:-1]  # Suppression de la première et de la dernière ligne

# Créer un graphique par tranche d'âge
fig_age_only = px.bar(
    df_age_avg_filtered,
    x="Âge",
    y="Taux",
    title="Taux de chômage moyen par tranche d'âge",
    labels={"Âge": "Tranche d'âge", "Taux": "Taux de chômage (%)"},
    color="Âge",  # Coloration automatique par tranche d'âge
    text="Taux",  # Afficher les valeurs sur les barres
)

# Personnalisation du graphique
fig_age_only.update_layout(
    font=dict(size=14, family="Arial"),
    plot_bgcolor="white",
    xaxis_title="Tranche d'âge",
    yaxis_title="Taux de chômage (%)",
)

# Ajouter des annotations aux barres pour plus de clarté
fig_age_only.update_traces(texttemplate='%{text:.2f}%', textposition='outside')

# Affichage du graphique
fig_age_only.show()
```

```python
# Filtrer les données pour une année donnée (par exemple 2020)
df_2020_sexe = df_long[df_long['Année'] == "2020"].groupby("Sexe", as_index=False)['Taux'].mean()

# Diagramme en secteurs
fig_sexe = px.pie(
    df_2020_sexe,
    values="Taux",
    names="Sexe",
    title="Répartition des taux de chômage par sexe en 2020",
    labels={"Taux": "Taux de chômage (%)", "Sexe": "Sexe"},
)

# Personnalisation
fig_sexe.update_traces(textinfo="percent+label")  # Ajouter les pourcentages et labels
fig_sexe.update_layout(
    font=dict(size=14, family="Arial"),
    plot_bgcolor="white",
)

fig_sexe.show()
```

```python
import plotly.express as px

# Filtrer les données pour les femmes de 15-24 ans dans df_long
df_femmes_15_24 = df_long[(df_long['Âge'] == '15-24 ans') & (df_long['Sexe'] == 'Femmes')]

# Créer un graphique interactif avec Plotly
fig = px.line(
    df_femmes_15_24,  # Utiliser le DataFrame filtré
    x="Trimestre",
    y="Taux",
    markers=True,
    title="Évolution du taux de chômage des femmes 15-24 ans (1975-2024)",
    labels={"Trimestre": "Trimestre", "Taux": "Taux de chômage (%)"},
)

# Personnalisation du graphique : Titre, axes, et style
fig.update_layout(
    title="Évolution du taux de chômage des femmes 15-24 ans (1975-2024)",
    xaxis_title="Trimestre",
    yaxis_title="Taux de chômage (%)",
    xaxis=dict(
        tickangle=45,  # Rotation des ticks
        tickmode="array",  # Pour gérer les labels manuellement
        tickvals=df_femmes_15_24['Trimestre'][::5],  # Afficher moins de ticks si besoin
    ),
    font=dict(family="Arial", size=14),
)

# Exporter le graphique en HTML interactif pour l'intégrer directement sur un site web
fig.write_html("graphique_interactif.html")

# Affichage du graphique
fig.show()
```

```python
import pandas as pd
import plotly.express as px

# Extraire l'année à partir de la colonne 'Trimestre' (par exemple '1975-T1' => '1975')
df_long['Année'] = df_long['Trimestre'].str[:4]

# Calculer la moyenne du taux de chômage par année et par sexe
df_moyenne_par_annee = df_long.groupby(['Année', 'Sexe', 'Âge'], as_index=False)['Taux'].mean()

# Créer un graphique interactif pour comparer les moyennes annuelles
fig = px.line(
    df_moyenne_par_annee,
    x="Année",
    y="Taux",
    color="Sexe",  # Séparation par sexe
    markers=True,
    title="Comparaison des taux de chômage annuels hommes et femmes 15-24 ans (1975-2024)",
    labels={"Année": "Année", "Taux": "Taux de chômage (%)", "Sexe": "Sexe"},
)

# Personnalisation du graphique pour améliorer la lisibilité
fig.update_layout(
    title="Comparaison des taux de chômage annuels hommes et femmes 15-24 ans (1975-2024)",
    xaxis_title="Année",
    yaxis_title="Taux de chômage (%)",
    xaxis=dict(
        tickangle=45,  # Rotation des ticks pour éviter la superposition
        tickmode="array",  # Contrôle des valeurs de l'axe X
        tickvals=df_moyenne_par_annee['Année'][::2],  # Afficher un tick tous les 2 ans
        showgrid=True  # Afficher les lignes de grille pour faciliter la lecture
    ),
    yaxis=dict(
        showgrid=True  # Ajouter les lignes de grille sur l'axe Y
    ),
    font=dict(family="Arial", size=14),  # Taille de la police
    legend=dict(
        title="Sexe",
        font=dict(size=12)
    ),
    plot_bgcolor="white",  # Fond blanc pour plus de clarté
    paper_bgcolor="white",  # Fond blanc de la page
)

# Affichage du graphique
fig.show()
```

```python
# Calculer la moyenne annuelle par tranche d'âge
df_age_moyenne_annuelle = df_long.groupby(['Année', 'Âge'], as_index=False)['Taux'].mean()

# Créer un graphique interactif par tranche d'âge
fig = px.line(
    df_age_moyenne_annuelle,
    x="Année",
    y="Taux",
    color="Âge",  # Séparation par tranche d'âge
    markers=True,
    title="Évolution des taux de chômage par tranche d'âge (1975-2024)",
    labels={"Année": "Année", "Taux": "Taux de chômage (%)", "Âge": "Tranche d'âge"},
)

# Personnalisation
fig.update_layout(
    xaxis_title="Année",
    yaxis_title="Taux de chômage (%)",
    xaxis=dict(tickangle=45, tickmode="array", tickvals=df_age_moyenne_annuelle['Année'][::2]),
    font=dict(family="Arial", size=14),
    plot_bgcolor="white",
    paper_bgcolor="white",
)

# Affichage
fig.show()
```

```python
# Calculer la moyenne annuelle par sexe et tranche d'âge
df_sexe_age_moyenne_annuelle = df_long.groupby(['Année', 'Sexe', 'Âge'], as_index=False)['Taux'].mean()

# Vérification des tranches d'âge disponibles
print("Tranches d'âge disponibles :", df_sexe_age_moyenne_annuelle['Âge'].unique())

# Exclure les tranches indésirables
ages_a_exclure = ['Taux de ch“mage de longue dur‚e']  # Ajustez ici si nécessaire

# Filtrer le DataFrame pour exclure les tranches indésirables
df_sexe_age_filtre = df_sexe_age_moyenne_annuelle[
    ~df_sexe_age_moyenne_annuelle['Âge'].isin(ages_a_exclure)
]

# Vérification des tranches après filtrage
print("Tranches d'âge après filtrage :", df_sexe_age_filtre['Âge'].unique())

# Créer un graphique interactif
fig = px.line(
    df_sexe_age_filtre,
    x="Année",
    y="Taux",
    color="Sexe",  # Comparaison entre hommes et femmes
    facet_col="Âge",  # Créer des sous-graphiques pour chaque tranche d'âge
    markers=True,
    title="Comparaison des taux de chômage hommes vs femmes par tranche d'âge (1975-2024)",
    labels={"Année": "Année", "Taux": "Taux de chômage (%)", "Sexe": "Sexe"},
)

# Personnalisation
fig.update_layout(
    xaxis_title="Année",
    yaxis_title="Taux de chômage (%)",
    font=dict(family="Arial", size=14),
    plot_bgcolor="white",
    paper_bgcolor="white",
)

# Affichage
fig.show()
```
